# Manipulação de dados com Pandas
### **Leitura**
- read_csv() ler arquivos CSV e outros dados estruturados
- sep: o separador de campo usado no arquivo CSV (,;|)
- decimal: separação da casa decimal (,.)

```python
import pandas as pd
df = pd.read_csv(r"data/Cancer_Data.csv", sep = ',', decimal='.')
df.head
```

- read_excel() para ler um arquivo Excel
- podemos especificar o nome da planilha que deseja ler passando o argumento sheet_name

```python
#!pip3 install openpyxl
df2 = pd.read_excel(r"data/fastfood.xlsx", sheet_name="fastfood")
df2.head()
```

### **Escrita**
- **to_csv()** escreve um DataFrame em um arquivo CSV
- **to_excel()** escreve um DataFrame em um arquivo Excel

```python
df2.to_excel("fastfood_new.xlsx")
```

# Índices 
- O index é uma sequência de rótulos que identifica exclusivamente cada linha do DataFrame. 
- O índice pode ser numérico, como um array de inteiros, ou pode ser uma sequência de rótulos, como uma lista de strings. 
- O método **set_index()** define uma ou mais colunas existentes do como o novo índice do DataFrame.
- O método **reset_index()** remove o índice atual e cria um novo índice numérico, começando do 0

- Utilize o argumento **drop=True** para remover o index.


```python
print(df2.index)
print(list(df2.index))
```
